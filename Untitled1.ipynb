```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')  # follow the link and authorize

# Example path to your file
df = pd.read_csv('/content/drive/MyDrive/dataset/covid.csv')
```

```python

# Convert numeric columns - remove commas from strings first
df['tot_cases'] = df['tot_cases'].astype(str).str.replace(',', '').astype(float)
df['new_cases'] = df['new_cases'].astype(str).str.replace(',', '').astype(float)

# Filter for Mississippi only
ms_data = df[df['state'] == 'MS'].copy().dropna()
ms_data = ms_data.sort_values('start_date')

print(f"Mississippi data points: {len(ms_data)}")
print(f"Current MS cases: {ms_data['tot_cases'].iloc[-1]:,.0f}")

class MississippiCOVIDSim:
    def __init__(self, data):
        self.data = data
        self.current_cases = data['tot_cases'].iloc[-1]

    def run_simulation(self, days=90, n_simulations=500):
        # Calculate realistic growth rates
        daily_growth = []
        for i in range(1, len(self.data)):
            prev_cases = self.data['tot_cases'].iloc[i-1]
            curr_cases = self.data['tot_cases'].iloc[i]
            if prev_cases > 0:
                growth = (curr_cases - prev_cases) / prev_cases
                daily_growth.append(growth)

        mean_growth = np.mean(daily_growth) if daily_growth else 0.01
        std_growth = np.std(daily_growth) if daily_growth else 0.005

        print(f"Current cases: {self.current_cases:,.0f}")
        print(f"Mean daily growth: {mean_growth:.4f}")
        print(f"Number of growth data points: {len(daily_growth)}")

        cases_proj = np.zeros((n_simulations, days))

        for sim in range(n_simulations):
            current = self.current_cases
            for day in range(days):
                growth = np.random.normal(mean_growth, std_growth)
                growth = np.clip(growth, -0.1, 0.2)
                current = current * (1 + growth)
                cases_proj[sim, day] = current

        self.cases_proj = cases_proj
        return cases_proj

    def plot_results(self):
        plt.figure(figsize=(12, 6))
        days = range(self.cases_proj.shape[1])

        mean = np.mean(self.cases_proj, axis=0)
        percentile_5 = np.percentile(self.cases_proj, 5, axis=0)
        percentile_95 = np.percentile(self.cases_proj, 95, axis=0)

        plt.plot(days, mean, 'b-', linewidth=2, label='Mean Projection')
        plt.fill_between(days, percentile_5, percentile_95, alpha=0.3, label='90% CI')

        plt.axhline(y=self.current_cases, color='red', linestyle='--',
                   label=f'Current: {self.current_cases:,.0f}')

        plt.xlabel('Days from Now')
        plt.ylabel('Total Cases')
        plt.title('Mississippi COVID-19 Cases Projection')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.ticklabel_format(style='plain', axis='y')
        plt.show()

# Run simulation
sim = MississippiCOVIDSim(ms_data)
sim.run_simulation()
sim.plot_results()
```
